# **Summer Olympics Data Analysis Assignment** **bold text**

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("summer.csv")
```

### **1. In how many cities Summer Olympics is held so far?**

```python
city = len(df['City'].unique())

print("Summer Olympics was held in " , city , " citis as so far")
```

### **2. Which sport is having most number of Gold Medals so far? (Top 5)**

```python
#value_counts ,Return a Series containing counts of unique values
#The resulting object will be in descending order so that the first element is the most frequently-occurring element.

df[df['Medal']=='Gold']['Sport'].value_counts().head().plot(kind = 'bar')
plt.xlabel("Sport")
plt.ylabel("Number of Gold Medals Won")
plt.title("Top 5 sports having most number of Gold")
plt.show()  
```

### **3. Which sport is having most number of medals so far? (Top 5)**

```python
df['Sport'].value_counts().head().plot(kind = 'bar' , color = 'orange')
plt.xlabel("Sport")
plt.ylabel("Number of Medals ")
plt.title("Top 5 sports having most number of Medals")
plt.show()
```

### **4. Which player has won most number of medals? (Top 5)**

```python
df['Athlete'].value_counts().head().plot(kind = 'bar' , color = 'green', figsize = (12, 5))
plt.xlabel("Athlete")
plt.ylabel("Number of Medals ")
plt.title("Top 5 Athlete having most number of Medals")
plt.xticks(rotation='horizontal')   # used to label x axis in horizontal directional
plt.show()
```

### **5. Which player has won most number Gold Medals of medals? (Top 5)**

```python
df[df['Medal']=='Gold']['Athlete'].value_counts().head().plot(kind = 'bar' , color = 'purple' , figsize = (10,5))
plt.xlabel("Athlete")
plt.ylabel("Number of Medals ")
plt.title("Top 5 Athlete having most number of Gold Medals")
plt.xticks(rotation='horizontal')   
plt.show()
```

### **6. In which year India won first Gold Medal in Summer Olympics?**

```python
data = df[df['Country']== 'IND']
#to.string is used to Whether to print index (row) labels.
result = data[data['Medal']== 'Gold']['Year'].sort_values(ascending = True).head(1).to_string(index = False)

print("India won its first summer olympic gold in the year '", result[1:],"'." )
```

### **7. Which event is most popular in terms on number of players? (Top 5)**

```python
print("Non - Unique Athlete")
#value_counts is uesd to return a Series containing counts of unique rows in the DataFrame.
df['Event'].value_counts().head().plot(kind = 'bar',color = 'grey' , figsize = (10, 6))
plt.title("Top 5 most popular events")
plt.xlabel("\n Event")
plt.ylabel(" Number of players (non - unique Athlete) ")
plt.xticks(rotation='horizontal')  
plt.show()
```

```python
print("Unique Athlete")
df_uni =[]
for event in df['Event'].unique():
    df_event = df[df['Event']==event]
    unique_athletes = len(df_event['Athlete'].unique())
    df_uni.append((event, unique_athletes))

df_uni = pd.DataFrame(df_uni , columns=[ 'Events' , 'Unique players']).sort_values(by='Unique players', ascending = False ).head()
#set_index used to set the DataFrame index using existing columns.
df_uni = df_uni.set_index('Events')
df_uni.plot(kind = 'bar' , color = 'brown', figsize = (12 ,8 ))
plt.title("Top 5 most popular events")
plt.xlabel("\n Event")
plt.ylabel(" Number of Unique players ")
plt.xticks(rotation='horizontal')  
plt.show()

```

### **8. Which sport is having most female Gold Medalists? (Top 5)**

```python
female =df[df['Gender']=='Women']
female = female[female['Medal']=='Gold']

female['Sport'].value_counts().head().plot(kind = 'bar' , color = 'pink' , figsize = (8 , 5))
plt.title("Top 5 sport with most female Gold Medalists ")
plt.xlabel("\n Sports")
plt.ylabel(" Number of Female Gold Medalists ")
plt.xticks(rotation='horizontal')  
plt.show()
```
